```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))
json = df[0].to_json(orient='records')
```

```python
df = pd.read_json(json)
df
```

```python
df = df[df['Borough']!='Not assigned']
df
```

```python
def not_assigned_to_borough(x):
    if x['Neighbourhood'] == 'Not assigned':
        x['Neighbourhood'] = x['Borough']
        return x
    else:
        return  x
```

```python
df = df.apply(not_assigned_to_borough, axis=1)
df
```

```python
df = df.groupby(['Borough'])['Neighbourhood'].apply(lambda x: ','.join(x)).reset_index()
df
```
